```python
import yfinance as yf
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from statsmodels.tsa.arima.model import ARIMA
import pmdarima as pm
from statsmodels.tsa.statespace.sarimax import SARIMAX
from sklearn.metrics import mean_squared_error
START_DATE = '2016-01-01'
END_DATE = '2024-05-29'
ASSET = ['AAPL', 'TSLA', 'MSFT', 'NVDA', 'GOOGL', 'AMZN', 'META']
prices_df = yf.download(ASSET, start=START_DATE, end=END_DATE)
print(f'Downloaded {prices_df.shape[0]} rows of data.')
print(prices_df.head())
big_7_returns = prices_df['Adj Close'].pct_change().dropna()
for stock in ASSET:
    print(f"Processing {stock}...")
    
    stock_returns = big_7_returns[stock]
    q_low = stock_returns.quantile(0.01)
    q_high = stock_returns.quantile(0.99)
    stock_returns_filtered = stock_returns[(stock_returns > q_low) & (stock_returns < q_high)]

    model_arima = pm.auto_arima(stock_returns_filtered, start_p=1, start_q=1, max_p=5, max_q=5, seasonal=False, d=1,
                                trace=True, error_action='ignore', suppress_warnings=True, stepwise=True)

    fc_arima, confint_arima = model_arima.predict(n_periods=len(stock_returns) + 12 * 4, return_conf_int=True)

    model_sarimax = SARIMAX(stock_returns_filtered, order=(2, 1, 2), seasonal_order=(1, 1, 1, 12))
    results_sarimax = model_sarimax.fit()

    fc_sarimax = results_sarimax.forecast(steps=len(stock_returns) + 12 * 4)

    mse_arima = mean_squared_error(stock_returns, fc_arima[:len(stock_returns)])
    mse_sarimax = mean_squared_error(stock_returns, fc_sarimax[:len(stock_returns)])
    print(f'MSE for ARIMA: {mse_arima}')
    print(f'MSE for SARIMAX: {mse_sarimax}')

    if mse_arima < mse_sarimax:
        selected_model = model_arima
        fc_final, confint_final = selected_model.predict(n_periods=12 * 10, return_conf_int=True)
    else:
        selected_model = results_sarimax
        forecast_result = selected_model.get_forecast(steps=12 * 10)
        fc_final = forecast_result.predicted_mean
        confint_final = forecast_result.conf_int()

    future_dates = pd.date_range(start=END_DATE, periods=12 * 10, freq='M')
    assert len(future_dates) == len(fc_final)

    predictions_df = pd.DataFrame({
        'Date': future_dates,
        'Forecasted Returns': fc_final
    })
    predictions_df.set_index('Date', inplace=True)

    predictions_df.to_csv(f'rolling_forecast_predictions_{stock}.csv', index=True)

    plt.figure(figsize=(12, 6))
    plt.plot(stock_returns.index, stock_returns, label='Actual Returns', color='blue')
    plt.plot(stock_returns.index, fc_arima[:len(stock_returns)], label='ARIMA Forecast', color='orange', linestyle='--')
    plt.plot(stock_returns.index, fc_sarimax[:len(stock_returns)], label='SARIMAX Forecast', color='green', linestyle='-.')
    plt.title(f'Actual Returns and Forecasted Returns for {stock}')
    plt.xlabel('Date')
    plt.ylabel('Returns')
    plt.legend()
    plt.show()
```
